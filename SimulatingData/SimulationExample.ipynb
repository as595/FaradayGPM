```python
# this embeds plots in the notebook
%matplotlib inline  
```

```python
import numpy as np  # for arrays
import pylab as pl  # for plotting
```

```python
from sim_data import *
```

```python
nu = 1e6*np.linspace(300.,3000.,256)
```

```python
sim = FaradaySim(nu,noise=0.0)
```

```python
phi_rg = 10.
P_rg = sim.FaradayThin(phi_rg)
```

```python
chi = 0.5*np.arctan2(P_rg.imag,P_rg.real)
chi*= (180./np.pi)   # convert radians to degrees
```

```python
fig, ax1 = pl.subplots()
ln1 = ax1.plot(sim.lam_sq, chi, 'b-',label=r"$\chi$")
ax1.set_xlabel(r"$\lambda^2$ [m$^2$]")
ax1.set_ylabel(r"$\chi$ [deg]")
ax1.set_ylim(-100, 100)

ax2 = ax1.twinx()
ln2 = ax2.plot(sim.lam_sq,np.absolute(P_rg),ls=':',c='r',label=r"$|P|$")
ln3 = ax2.plot(sim.lam_sq,P_rg.real,ls='--',label=r"$Q_{\rm fg}$")
ax2.set_ylabel("Flux [Jy]")
ax2.set_ylim(-1.2, 1.2)

# figure legend:
lns = ln1+ln2+ln3
labs = [l.get_label() for l in lns]
ax2.legend(lns, labs, loc=1)

fig.tight_layout()
pl.savefig("Fig1.png")
pl.show()
```
